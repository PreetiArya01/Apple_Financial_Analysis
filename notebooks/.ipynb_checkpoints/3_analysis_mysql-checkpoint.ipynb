### Profit Margin Trend

### Gross Profit Margin (GPM) Analysis  

- **Definition**:  
  Gross Profit Margin shows how much profit a company makes after deducting the **cost of goods sold (COGS)** from revenue. It reflects the efficiency of production and pricing.  

  \[
  \text{Gross Profit Margin} = \frac{\text{Revenue – COGS}}{\text{Revenue}} \times 100
  \]

- **Purpose**:  
  - Indicates how well Apple controls its production costs relative to sales.  
  - A higher margin means Apple retains more from each dollar of sales to cover operating expenses, R&D, and profits.  
  - Useful for comparing performance over years or against competitors.  

- **Interpretation**:  
  -  **High GPM (>40% in tech/consumer electronics)**: Strong product pricing power and cost efficiency.  
  -  **Declining GPM**: Rising production costs, pricing pressure, or supply chain challenges.  
  -  **Stable GPM**: Consistency in Apple’s cost management and pricing strategy.  

- **Analysis in Apple's Context**:  
  - Apple is known for maintaining high gross margins due to its premium pricing and efficient supply chain.  
  - Tracking GPM over the years reveals the impact of new product launches, component costs (e.g., chips, displays), and global supply issues.

### Debt-to-Equity (D/E) Ratio Analysis  

- **Definition**:  
  The Debt-to-Equity ratio measures a company’s financial leverage by comparing its total liabilities to shareholders’ equity.  
  \[
  \text{Debt-to-Equity Ratio} = \frac{\text{Total Debt}}{\text{Shareholders' Equity}}
  \]

- **Purpose**:  
  - Helps investors and analysts understand how Apple finances its operations — through debt or equity.  
  - A higher ratio means the company is relying more on debt, which may increase risk if earnings fall.  
  - A lower ratio indicates more conservative financing with less reliance on debt.  

- **Interpretation**:  
  -  **High D/E Ratio (>1)**: Apple is financing growth primarily with debt. This can magnify returns in good times but increases financial risk.  
  -  **Low D/E Ratio (<1)**: Apple is less leveraged, suggesting financial stability but possibly slower growth.  
  -  **Balanced D/E Ratio (~1)**: A healthy mix of debt and equity financing.  


- **Analysis in Apple's Context**:  
  - Observing the D/E trend over multiple years highlights whether Apple is increasing leverage to fund innovation, acquisitions, or shareholder payouts.  
  - Sharp changes in the ratio should be cross-checked with **Cash Flow** and **Interest Expense** trends to assess sustainability.

### Net Income Trend Analysis

- **Definition**:  
  Net Income (also called "bottom line") represents the company's total profit after accounting for all expenses, taxes, and costs.  
  It is one of the most critical indicators of financial health and profitability.  

- **Analysis**:  
  By plotting the trend of Apple's Net Income over the years, we can evaluate:  
  - Whether the company is consistently growing profits.  
  - The impact of external factors (e.g., market conditions, product launches, or global events) on profitability.  
  - Any decline or volatility that may require deeper investigation into costs or revenues.  

- **Interpretation**:  
  -  An **upward trend** indicates Apple is increasing its profitability and managing costs effectively.  
  -  A **downward trend** may signal rising expenses, declining sales, or market challenges.  
  -  Fluctuations suggest cyclical business performance or external shocks.  


```python
important_cashflow = [
    "Operating Cash Flow",
    "Investing Cash Flow",
    "Financing Cash Flow",
    "Free Cash Flow"
]

filtered_cf = cashflow_analysis_df[cashflow_analysis_df["Metric"].isin(important_cashflow)]

plt.figure(figsize=(10,6))
for metric in important_cashflow:
    subset = filtered_cf[filtered_cf["Metric"] == metric]
    plt.plot(subset["Year"], subset["Value"], marker="o", label=metric)

plt.title("Apple Cash Flow Overview")
plt.xlabel("Year")
plt.ylabel("USD (Billions)")
plt.legend()
plt.grid(True)
plt.show()

```

### Interpretation
- Operating Cash Flow stays positive → Apple generates strong core earnings.
- Investing Cash Flow is negative → Apple spends heavily on R&D and capital assets.
- Financing Cash Flow is negative → Apple repays debt and buys back shares.
- Free Cash Flow is positive → Apple has cash left after investments, showing financial health.

```python
roe_analysis_cleaned = roe_analysis_df.drop_duplicates()
```

```python
# ROE 
plt.figure(figsize=(8,5))
plt.plot(roe_analysis_cleaned["Year"], roe_analysis_cleaned["ROE"], marker="o", linewidth=2)

plt.title("Apple Return on Equity (ROE)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("ROE (%)")
plt.grid(True)
plt.show()
```

### ROE Analysis
- ROE measures how effectively Apple generates profit from shareholders' equity.  
- A consistently high ROE (>20%) indicates strong profitability and efficient capital use.  
- If ROE declines, it may signal rising equity without proportional income growth.
